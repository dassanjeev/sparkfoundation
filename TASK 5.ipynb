### Importing Required packages And DataSet

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("Indian Premier League/deliveries.csv")
data.head()
```

```python
data.shape
```

```python
sns.heatmap(data.isnull(),cmap="coolwarm")
plt.show()
```

```python
data.columns
```

```python
data["batting_team"].value_counts()
```

```python
sns.barplot(x="batting_team",y="wide_runs",data=data)
plt.xticks(rotation=90)
plt.show()
```

## It seems Kochi Tuskers Kerala had a highest rate of wide per ball

```python
sns.barplot(x="batting_team",y="extra_runs",data=data)
plt.xticks(rotation=90)
plt.show()
```

## Kochi Tuskers Kerala also gave a high number of extra runs

```python
data[6*20:6*40]
```

```python
MI=data[data["batting_team"]=="Mumbai Indians"]
```

```python
np.average(MI[MI["batsman"]=="SR Tendulkar"]["total_runs"])
```

```python
plt.figure(figsize=(16,4))
sns.barplot(x="batsman",y="total_runs",data=MI)
plt.xticks(rotation=90)
plt.show()
```

## This is the run rate per ball of the Batsman of Mumbai Indian
### Note: the Black line is the variablity of the runrate 
### RG Shrama , IC Butler , AT Rayudu , KH Pandaya , HH Pandaya and Pollard are one of the important batsman as they have a very high runrate 

```python
plt.figure(figsize=(16,4))
sns.barplot(x="over",y="total_runs",data=MI,color="lightblue")
plt.xticks(rotation=90)
plt.show()
```

## Batting Strategy of Mumbai indian seems like they tend to have higher runrate at the last overs compared to the initial overs


```python
plt.figure(figsize=(16,4))
sns.barplot(x="batting_team",y="total_runs",data=data,palette="Blues")
plt.xticks(rotation=90)
plt.show()
```

## Royal Challengers Bangalore, Mumbai Indian,  Kolkata knight Riders, Chennai Super Kings and Delhi Capitals seems to have better performance than others

## Virat Kohli seems to have a very high run rate at the end and also the variablity of virate per over is less

## Virat Kohli has a high potential of higting for a greater number in 1st , 2nd 4th and 5th ball compared to 3rd and 6th ball

```python
plt.figure(figsize=(16,4))
sns.barplot(x="over",y="total_runs",data=data[data["batsman"]=="CH Gayle"],color="orange")
plt.xticks(rotation=90)
plt.show()
```

## Chris Gayle as a batsman has a high runrate but has a lot of variablility in his run rate and this can be a problem for the team

```python
plt.figure(figsize=(16,4))
sns.barplot(x="over",y="total_runs",data=data[data["batsman"]=="HH Pandya"],color="lightblue")
plt.xticks(rotation=90)
plt.show()
```

## HH Pandya has a lot of variablity in the intial over but tends to get stable in the last overs 


```python
plt.figure(figsize=(16,4))
sns.barplot(x="over",y="total_runs",data=data[data["batsman"]=="MS Dhoni"],color="yellow")
plt.xticks(rotation=90)
plt.show()
```

## Dhoni can be seen as one of the stable players where his runrate increases with overs and has a very low variablity 
### It can be seen that Dhoni doesnt have a runrate at the initial overs ... is the fact that he goes later in the batting position of 4 or 5

### All teams are equally go but on terms of runrate Delhi Capitals, Royal Challengers Bangalore, Mumbai Indian, Chennai Super Kings and Kolkata Knight Riders are one of the stable teams 
### And Dhoni and Virat Kohli are one of the important players with low variability and high Score
